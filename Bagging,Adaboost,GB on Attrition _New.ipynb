```python
import pandas as pd
import numpy as np
from sklearn.metrics import *
from sklearn.ensemble import AdaBoostClassifier
from sklearn.preprocessing import *
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
attrition=pd.read_csv('C:\\Users\\ABC\\Downloads\\attrition.csv')
```

```python
#Base Gradient Boosting Classifier
gb=GradientBoostingClassifier()
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x=attrition.drop('Attrition',axis=1)
```

```python
y=attrition.Attrition
```

```python
x
```

```python
x_num=x.select_dtypes('number')
x_cat=x.select_dtypes('object')
```

```python
from sklearn.preprocessing import scale
```

```python
x_num_scaled=pd.DataFrame(scale(x_num),columns=x_num.columns)
```

```python
x_num_scaled
```

```python
x_cat_dummies=pd.get_dummies(x_cat)
```

```python
x_cat_dummies
```

```python
x_processed=pd.concat([x_num_scaled,x_cat_dummies],axis=1)
```

```python
x_processed
```

```python
x_train,x_test,y_train,y_test=tts(x_processed,y,stratify=y,random_state=365,test_size=0.2)
```

```python
gb.fit(x_train,y_train)
```

```python
pd.DataFrame(data=[accuracy_score(y_train,gb.predict(x_train)), recall_score(y_train, gb.predict(x_train)),
                   precision_score(y_train, gb.predict(x_train)), roc_auc_score(y_train, gb.predict(x_train)),
                   f1_score(y_train,gb.predict(x_train))], 
             index=["accuracy", "recall", "precision", "roc_auc_score","f1_score"])
```

```python
pd.DataFrame(data=[accuracy_score(y_test,gb.predict(x_test) ), recall_score(y_test, gb.predict(x_test)),
                   precision_score(y_test, gb.predict(x_test)), roc_auc_score(y_test, gb.predict(x_test)),
                   f1_score(y_test,gb.predict(x_test))], 
             index=["accuracy", "recall", "precision", "roc_auc_score","f1_score"])
```

```python
parameters={'loss':['log_loss','deviance','exponential'],'learning_rate':[0.001,0.01,0.1,1,10],'criterion':['friedman_mse',"squared_error"],
           'min_samples_leaf':[2,5,8,11,14],'min_samples_split':[2,5,8,11,14],'max_depth':[5,10,15,20,25],'max_features':['auto','sqrt','log2'],
           'max_leaf_nodes':[2,4,6,8,10],'max_features':['auto','sqrt','log2']}
```

```python
gridr=RandomizedSearchCV(estimator=gb,param_distributions=parameters,scoring='f1',n_jobs=4,cv=5,return_train_score=True
                       ,refit='f1',n_iter=10000)
```

```python
gridr.fit(x_train,y_train)
```

```python
gridr.best_score_
```

```python
gridr.best_params_
```

```python
gb2=GradientBoostingClassifier(**gridr.best_params_)
```

```python
gb2.fit(x_train,y_train)
```

```python
pd.DataFrame(data=[accuracy_score(y_test,gb2.predict(x_test) ), recall_score(y_test, gb2.predict(x_test)),
                   precision_score(y_test, gb2.predict(x_test)), roc_auc_score(y_test, gb2.predict(x_test)),
                   f1_score(y_test,gb2.predict(x_test))], 
             index=["accuracy", "recall", "precision", "roc_auc_score","f1_score"])
```

```python
#optimizing threshold on b
def threshold1(n=1000):
    th=np.linspace(0,1,n)
    f1=[]
    y_pred=b.predict_proba(x_test)
    for i in th:  
        y_pred1=np.where(y_pred[:,1]>i,1,0)
        f1.append(f1_score(y_test,y_pred1))
    return th[f1.index(max(f1))]
```

```python
threshold1()
```

```python
y_pred_testb=np.where(b.predict_proba(x_test)[:,1]>threshold1(),1,0)
```

```python
pd.DataFrame(data=[accuracy_score(y_test,y_pred_testb ), recall_score(y_test, y_pred_testb),
                   precision_score(y_test, y_pred_testb), roc_auc_score(y_test, y_pred_testb),
                   f1_score(y_test,y_pred_testb)], 
             index=["accuracy", "recall", "precision", "roc_auc_score","f1_score"])
```
